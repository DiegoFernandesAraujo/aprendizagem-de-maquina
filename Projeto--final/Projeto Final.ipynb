## <center>Resolução de entidades com aprendizagem de máquina</center>

## Leitura e pré-processamento dos dados

Para o experimento foi utilizada uma base de dados onde cada registro refere-se a um par de CDs musicais, a qual foi gerada a partir da aplicação do produto cartesiano sobre todos os registros de uma base de dados disponibilizada [publicamente](http://www13.hpi.uni-potsdam.de/ﬁleadmin/user_upload/fachgebiete/naumann/projekte/dude/cd.csv)
que contém 9763 registros de CDs musicais com atributos como título, artista, ano de lançamento
e estilo musical.

Os registros da base de dados gerada representam os pares resultantes do produto cartesiano aplicado sobre a base de dados de CDs musicais. Cada campo que compõe os registros armazena o valor de similaridade existente entre os valores de cada atributo (por exemplo, nome ou nascimento) para cada par. Esses valores de similaridade estão no intervalo $[0,1] \in \mathbb{R}$, de forma que quanto mais próximo de $1$, mais similar são os dados daquele campo para o par.

Dos 47.653.502 pares gerados apenas 299 entidades são duplicatas, segundo o _Gold Standard_ disponibilizado junto com a base de dados.

Devido a base de dados resultante possuir muitos _GigaBytes_ de informação, por limitações técnicas referentes a memória, para essa tarefa foi utilizada uma amostra dos dados com 1.048.576 de pares de não-duplicatas e 299 duplicatas.

```python
dup = pd.read_csv("DuplicatasCD.csv", index_col=False, names=('title', 'artist', 'track01', 'track02', 'track03', 'duplicata'), header = None)
noDup = pd.read_csv("NaoDuplicatasCD.csv", index_col=False, names=('title', 'artist', 'track01', 'track02', 'track03', 'duplicata'), header = None)
dataFrames = [dup, noDup]
all = pd.concat(dataFrames)
```

### Extração de features

A seguir é adicionada uma nova coluna (*feature*) que pode ajudar na classificação a ser realizada pelo modelo.
Aqui presume-se que a relação entre artista e título possa ter um impacto maior do que quando considerados separadamente.

```python
all['artist-title'] = all['artist'] * all['title']
cols = list(all.columns.values)
cols.pop(cols.index('duplicata'))
all = all[cols+['duplicata']]
all.head()
```

### Estatísticas dos dados

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
print(all.describe())
```

Como os valores de similaridade armazenados em cada campo pertencem ao intervalo $[0,1]$, os dados não necessitam de qualquer tipo de normalização.

```python
#Separação do conjunto X do conjunto y
X = all.loc[:,'title':'artist-title']
y = all.duplicata
```

```python
from collections import Counter
print("Quantidade de itens em y: {}".format(Counter(y)))
```

```python
dup = [i for i in y if i == True]
percentage = float(dup.count(True)/float(len(y)))
print("{0}% de duplicatas.".format(str(percentage)[:6]))
```

Como dito anteriormente, há um enorme número de não-duplicatas para uma pequena quantidade de duplicatas. Esse grande desbalanceamento pode afetar negativamente o treinamento do modelo.

## Reamostragem

Uma alternativa par lidar com desbalanceamento de classes em problemas de classificação que utilizam algoritmos de aprendizagem de máquina é fazer uso de técnicas de reamostragem de dados.

Os métodos utilizados para reamostragem podem ser de dois tipos:

* ** *Undersampling***, onde são deletadas instâncias da classe majoritária (nesse caso, entidades não-duplicatas); e

* ** *Oversampling***, onde são criadas cópias das instâncias da classe minoritária;

Aqui foi utilizado um método de reamostragem sintético denominado *SMOTE* (*Synthetic Minority Oversampling Technique*) (CHAWLA et al., 2002),  o qual cria amostras sintéticas da classe minoritária com base nos segmentos de reta que ligam cada exemplo positivo $x_i$ aos seus $k$ vizinhos mais
próximos (CASTRO e BRAGA, 2011), conforme ilustra a figura abaixo.

<img src="smote.png" style="width: 300px;"/>

Antes de aplicar a reamostragem ao conjunto de dados, iremos dividir os conjuntos de treino e teste, mantendo 80% dos exemplos para o primeiro e 20% para o último.

```python
from sklearn import model_selection

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.20, random_state=2)
```

Aplicação de reamostragem *SMOTE* com taxa de 50% indicando que a quantidade de exemplos da classe minoritária se igualará à quantidade correspondente à metade dos exemplos da classe majoritária.

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

overSampling = SMOTE(ratio = 0.5, random_state = 1)
X_train_res, y_train_res = overSampling.fit_sample(X_train, y_train)

print("Distribuição de classes antes de realizar a reamostragem {}".format(Counter(y_train)))
print("Distribuição de classes depois de realizar a reamostragem {}".format(Counter(y_train_res)))
```

## Construção dos modelos

Com os dados tratados, vamos construir modelos de classificação com o intuito de predizer se um dado par de entidades é duplicado ou não.

### Validação cruzada

Para selecionar um modelo adequado será utilizada validação cruzada *k-fold* que consiste em separar os dados em *k* conjuntos diferentes de forma que seja possível executar *k* experimentos de treino e validação, cada vez mantendo um conjunto diferente para avaliação. Ao fim é calculada a média dos valores de performance de validação dos *k* experimentos para que se possa estimar o erro de generalização de cada modelo. Na figura abaixo é apresentado o processo para $k=4$.

<img src="k-fold.png" style="width: 400px;"/>

#### *Stratified kfold*

A fim de garantir que cada *fold* seja representativo de todos estratos dos dados será usado aqui *Stratified kfold*, o qual a partir de um processo interno verifica se cada *fold* contém aproximadamente o mesmo número de exemplos de cada classe, reorganizando os dados aleatoriamente caso essa condição não seja satisfeita.

#### Métrica de avaliação (Função de custo)

Como o problema aqui abordado consiste em identificar os casos positivos de duplicatas corretamente, além de encontrar todos os existentes, isso envolve a aplicação das métricas *precision* e *recall*, respectivamente.

<img src="metricas.png" style="width: 200px;"/>

Entretanto, como essas métricas são conflitantes, ou seja, quando uma alcança bons valores, a outra tende a apresentar um desempenho mais baixo e vice-versa, será utilizada a métrica $F1$. Essa métrica calcula a média harmônica entre *precision* e *recall*, favorecendo, assim, classificadores que são fortes em ambas.

$$F1 = \frac{2 * recall * precision}{recall + precision}$$

Abaixo é executada validação cruzada sobre 4 modelos de aprendizagem de máquina para classificação, a fim de selecionar um a ser utilizado para predizer à classe a qual pertence um dado par de entidades.

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold

modelos = []
modelos.append(('LR', LogisticRegression()))
modelos.append(('KNN', KNeighborsClassifier()))
modelos.append(('DTC', DecisionTreeClassifier()))
modelos.append(('SVM', SVC()))

seed = 500

# Avaliação de cada modelo por vez
resultados = []
nomes = []
for nome, modelo in modelos:
    kfold = StratifiedKFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(modelo, X_train, y_train, cv=kfold, scoring='f1')
    msg = "%s: %f (%f)" % (nome, cv_results.mean(), cv_results.std())
    print(msg)
```

Pelo apresentado não percebemos diferença significativa entre os modelos.

O modelo **DTC** correspondente a uma árvore de decisão, considerando apenas a média de $F1$, apresentou melhor resultado. Pelo fato dos resultados desse modelo específico ser de fácil interpretação, o mesmo será utilizado daqui em diante.

```python
dtc = DecisionTreeClassifier()
```

## Tunando o modelo

Como o modelo utilizado apresenta as configurações _default_ não há limitação para o tamanho da árvore a ser gerada, o que tende a gerar *overfitting*, assim iremos realizar validação cruzada a fim de identificar a melhor seleção de hiperparâmetros para tunar nosso modelo.

Para que isso seja possível, são selecionados intervalos de valores de cada hiperparâmetro a serem explorados. Com isso é realizado um produto cartesiano entre os intervalos, o que leva a uma busca exaustiva pela melhor combinação de hiperparâmetros, de maneira que ao término do processo é retornada a combinação que resultou no melhor valor de performance durante a validação. 

```python
#Quais os hiperparâmetros utilizados por uma árvore de decisão?
dtc.get_params()
```

De posse dos hiperparâmetros iremos considerar os seguintes para a validação:

* ** *criterion***, que corresponde à medidade de qualidade utilizada para dividir a árvore;

* ** *max_depth***, que estabelece a profundidade máxima da árvore;

* ** *class_weight***, que determina peso às classes. Caso nenhum peso seja atribuído, todas as classes assumem peso 1;

* ** *min_samples_split***, que indica o número mínimo de amostras requeridas para dividir um nó interno.

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{"criterion": ['gini', 'entropy'],"max_depth": [2,3,4,5,6,7],
              "splitter": ["best","random"],"class_weight": ["balanced",None],
              "min_samples_split": [2, 5, 10]}]
    
score = 'f1'

print("# Tunando hiperparâmetros para %s" % score)
print()

grid = GridSearchCV(dtc, param_grid, cv=5, scoring='%s_macro' % score)
grid.fit(X_train, y_train)

print("Melhor conjunto de hiperparâmetros encontrado:")
print()
print(grid.best_params_)
print()
print("Grade de valores encontrados:")
print()
medias = grid.cv_results_['mean_test_score']
desvios = grid.cv_results_['std_test_score']
for media, desvio, params in zip(medias, desvios, grid.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (media, desvio * 2, params))
    print()
```

```python
dtc2 = DecisionTreeClassifier(**grid.best_params_, random_state=500)

cv_results2 = model_selection.cross_val_score(dtc2, X_train, y_train, cv=kfold, scoring='f1')
msg = "%s com hiperparâmetros tunados: F1 = %f com desvio padrão = %f" % ('DTC', cv_results2.mean(), cv_results2.std())
print(msg)
```

O resultado de $F1$ é semelhante àquele da configuração _default_. Como essa nova árvore tem profundidade 7 e, portanto, mais simples, ela será utilizada para predizer os dados de teste.

## Predição sobre os dados de teste

```python
dtc2.fit(X_train, y_train)
predicoes = dtc2.predict(X_test)
```

### Resultados

#### Matriz de confusão

Uma matriz de confusão $C$ confronta os valores preditos pelo modelo com os valores reais, de modo que, para uma classificação binária, $C_{0,0}$ computa os casos verdadeiros negativos, $C_{1,0}$ os falsos negativos, $C_{1,1}$ os verdadeiros positivos e $C_{0,1}$ os falsos positivos, conforme exposto na figura a seguir.

<img src="matriz.png" style="width: 400px;"/>

```python
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, predicoes)

print(matriz)
```

Pela matriz acima percebe-se que o modelo foi capaz de predizer corretamente 48 duplicatas, errando o resultado apenas 5 vezes tanto para a predição dessas quanto para as não-duplicatas.

#### *Precision*, *Recall* e *F1*

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predicoes))
```

Pelos dados apresentados acima, observa-se que o modelo construído apresentou-se bastante eficaz, quse alcançando perfeição na predição de não-duplicatas e apresentando ótimo valor de $F1$ para a predição de duplicatas, capaz de identificar 91% delas com 91% de precisão.
